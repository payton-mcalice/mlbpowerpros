```python
#for a new player, randomly assign their player stats at max potential
def max_stats():
    traj = (randint(25, 50, 1)[0]) / 10
    contact = (randint(49, 160, 1)[0]) / 10
    power = (randint(259, 2560, 1)[0]) / 10
    run = (randint(49, 160, 1)[0]) / 10
    arm = (randint(49, 160, 1)[0]) / 10
    field = (randint(49, 160, 1)[0]) / 10
    error = (randint(49, 160, 1)[0]) / 10
    con_pot = round((np.floor(contact) / 15) * 100)
    bat_pot = round((((np.floor(traj) / 4) + (np.floor(power)/ 255)) / 2) * 100)
    run_pot = round((np.floor(run) / 15) * 100)
    field_pot = round((((np.floor(arm) / 15) + (np.floor(field) / 15) + (np.floor(error)/ 15)) / 3) * 100)
    potential = round(((con_pot / 4) + (bat_pot / 4) + (run_pot / 4) + (field_pot / 4)))
    return traj, contact, power, run, arm, field, error, con_pot, bat_pot, run_pot, field_pot, potential
```

```python
#assign a new player their positions and skill-level at said positions
def create_positions():
    Positions = {1: "N/A", 2: "C", 3: "1B", 4: "2B", 5: "3B", 6: "SS", 7: "LF", 8: "CF", 9: "RF"}
    Pos1 = randint(2, 10, 1)[0]
    Pos2 = 1
    Pos3 = 1
    Flip1 = randint(0, 2, 1)[0]
    Flip2 = randint(0, 2, 1)[0]
    if Flip1 == 0:
        Pos2 = randint(2, 10, 1)[0]
        if Pos1 == Pos2:
            Pos2 = 1
    if Flip1 == 0 and Flip2 == 0:
        Pos3 = randint(2, 10, 1)[0]
        if Pos1 == Pos3 or Pos2 == Pos3:
            Pos3 = 1
    Pos1 = Positions[Pos1]
    Pos2 = Positions[Pos2]
    Pos3 = Positions[Pos3]

    Grades = {1: "E", 2: "D", 3: "C", 4: "B", 5: "A"}
    P2G = randint(1, 6, 1)[0]
    P3G = randint(1, 6, 1)[0]
    if Pos2 != "N/A":
        Pos2 = Pos2 + "(" + Grades[P2G] + ")"
    if Pos3 != "N/A":
        Pos3 = Pos3 + "(" + Grades[P3G] + ")"
    return Pos1, Pos2, Pos3
```

```python
#assign a new player an age as well as a development arc
def age_etc():
    age = randint(21, 36, 1)[0]
    retired_age = randint(35, 41, 1)[0]
    contract = 0
    if age < 30:
        contract = randint(1, 11, 1)[0]
    if age >= 30:
        contract = randint(1, 7, 1)[0]
        
    peak = randint(24, 31, 1)[0]
    develop = randint(20, 23, 1)[0]
    level = randint(peak + 1, 33, 1)[0]
    decline = level + randint(2, 6, 1)[0]
    tank = randint(32, 42, 1)[0]
    while tank <= decline:
        tank = tank + 1
    return age, develop, peak, level, decline, tank, retired_age, contract
```

```python
#determine how much of a player's max potential is reached for each skill
def skill_percents(age, develop, peak, level, decline, tank):
    traj_per = 0
    con_per = 0
    pow_per = 0
    run_per = 0
    arm_per = 0
    fld_per = 0
    err_per = 0
    if age <= develop:
        percents = randint(50, 77, 7)
        traj_per = percents[0]
        con_per = percents[1]
        pow_per = percents[2]
        run_per = percents[3]
        arm_per = percents[4]
        fld_per = percents[5]
        err_per = percents[6]
    if age > develop and age <= peak:
        percents = randint(63, 90, 7)
        traj_per = percents[0]
        con_per = percents[1]
        pow_per = percents[2]
        run_per = percents[3]
        arm_per = percents[4]
        fld_per = percents[5]
        err_per = percents[6]
    if age > peak and age <= level:
        percents = randint(76, 101, 7)
        traj_per = percents[0]
        con_per = percents[1]
        pow_per = percents[2]
        run_per = percents[3]
        arm_per = percents[4]
        fld_per = percents[5]
        err_per = percents[6]
    if age > level and age <= decline:
        percents = randint(70, 101, 7)
        traj_per = percents[0]
        con_per = percents[1]
        pow_per = percents[2]
        run_per = percents[3]
        arm_per = percents[4]
        fld_per = percents[5]
        err_per = percents[6]
    if age > decline and age <= tank:
        percents = randint(68, 89, 7)
        traj_per = percents[0]
        con_per = percents[1]
        pow_per = percents[2]
        run_per = percents[3]
        arm_per = percents[4]
        fld_per = percents[5]
        err_per = percents[6]
    if age > tank:
        percents = randint(50, 77, 7)
        traj_per = percents[0]
        con_per = percents[1]
        pow_per = percents[2]
        run_per = percents[3]
        arm_per = percents[4]
        fld_per = percents[5]
        err_per = percents[6]
    return traj_per, con_per, pow_per, run_per, arm_per, fld_per, err_per
```

```python
#randomly generate a hot-cold batter zone
def hot_cold():
    hc = {1: "C", 2: "-", 3: "-", 4: "H"}
    upleft = hc[randint(1, 5, 1)[0]]
    upmid = hc[randint(1, 5, 1)[0]]
    upright = hc[randint(1, 5, 1)[0]]
    midleft = hc[randint(1, 5, 1)[0]]
    midmid = hc[randint(1, 5, 1)[0]]
    midright = hc[randint(1, 5, 1)[0]]
    downleft = hc[randint(1, 5, 1)[0]]
    downmid = hc[randint(1, 5, 1)[0]]
    downright = hc[randint(1, 5, 1)[0]]
    return upleft + upmid + upright + midleft + midmid + midright + downleft + downmid + downright
```

```python
#determine how skilled a pitcher is in different stats
def skill_percents_p(age, develop, peak, level, decline, tank):
    speed_per = 0
    control_per = 0
    stamina_per = 0
    break_per = 0
    if age <= develop:
        percents = randint(55, 77, 4)
        speed_per = percents[0]
        control_per = percents[1]
        stamina_per = percents[2]
        break_per = percents[3]
    if age > develop and age <= peak:
        percents = randint(68, 90, 4)
        speed_per = percents[0]
        control_per = percents[1]
        stamina_per = percents[2]
        break_per = percents[3]
    if age > peak and age <= level:
        percents = randint(81, 101, 4)
        speed_per = percents[0]
        control_per = percents[1]
        stamina_per = percents[2]
        break_per = percents[3]
    if age > level and age <= decline:
        percents = randint(75, 101, 4)
        speed_per = percents[0]
        control_per = percents[1]
        stamina_per = percents[2]
        break_per = percents[3]
    if age > decline and age <= tank:
        percents = randint(73, 89, 4)
        speed_per = percents[0]
        control_per = percents[1]
        stamina_per = percents[2]
        break_per = percents[3]
    if age > tank:
        percents = randint(55, 77, 4)
        speed_per = percents[0]
        control_per = percents[1]
        stamina_per = percents[2]
        break_per = percents[3]
        
    if speed_per < 75:
        speed_per = 75
        
    return speed_per, control_per, stamina_per, break_per
```

```python
#find the standard deviation using a given mean
def era_stdev(mean, eras):
    for era in eras:
        if era == 42069:
            eras.remove(era)
    size = len(eras)
    sum_dev = 0
    for era in eras:
        sum_dev += (era - mean) ** 2
    stdev = ((sum_dev)/size) ** 0.5
    return stdev
```

```python
#calculate the Z-PITCH statistic of a player
def zpitcher(season, teams, pitcher_stats):    
    hot_robot_pitchers = pitcher_stats[(pitcher_stats['TEAM'].isin(teams))]
    hot_robot_pitchers = hot_robot_pitchers[hot_robot_pitchers["SEASON"] == season].reset_index()
    hot_robot_pitchers["IPfloat"] = 1
    for i, hot_robot_pitcher in hot_robot_pitchers.iterrows():
        hot_robot_pitchers.at[i, "IPfloat"] = float(hot_robot_pitcher["IP"]) * 100
    hot_robot_pitchers["ERfloat"] = 1
    for i, hot_robot_pitcher in hot_robot_pitchers.iterrows():
        hot_robot_pitchers.at[i, "ERfloat"] = float(hot_robot_pitcher["ER"]) * 100
    hot_robot_pitchers["ERAfloat"] = 1
    for i, hot_robot_pitcher in hot_robot_pitchers.iterrows():
        if hot_robot_pitcher["ERA"] == "inf" or hot_robot_pitcher["ERA"] == "N/A":
            hot_robot_pitchers.at[i, "ERAfloat"] = 42069
        else:
            hot_robot_pitchers.at[i, "ERAfloat"] = float(hot_robot_pitcher["ERA"]) * 100
    max_ip = max(hot_robot_pitchers["IPfloat"])
    hot_robot_mean_ip = hot_robot_pitchers["IPfloat"].mean()
    hot_robot_sum_ip = hot_robot_pitchers["IPfloat"].sum()
    hot_robot_sum_er = hot_robot_pitchers["ERfloat"].sum()
    hot_robot_mean_era = (hot_robot_sum_er / hot_robot_sum_ip) * 600
    hot_robot_stdev_ip = hot_robot_pitchers["IPfloat"].std()    
    hot_robot_stdev_era = era_stdev(hot_robot_mean_era, hot_robot_pitchers["ERAfloat"].tolist())
    hot_robot_pitchers["ZERA"] = 0
    for i, hot_robot_pitcher in hot_robot_pitchers.iterrows():
        hot_robot_pitchers.at[i, "ZERA"] = ((hot_robot_mean_era - hot_robot_pitcher["ERAfloat"]) / hot_robot_stdev_era) * 100
    hot_robot_pitchers["ZIP"] = 0
    for i, hot_robot_pitcher in hot_robot_pitchers.iterrows():
        hot_robot_pitchers.at[i, "ZIP"] = ((hot_robot_pitcher["IPfloat"] - hot_robot_mean_ip) / hot_robot_stdev_ip) * 42.5
    hot_robot_pitchers["ZPITCH"] = 0
    
    buff = (((1/3) - hot_robot_mean_ip) / hot_robot_stdev_ip) * -42.5
    for i, hot_robot_pitcher in hot_robot_pitchers.iterrows():
        if hot_robot_pitcher["ERA"] == "inf":
            hot_robot_pitchers.at[i, "ZPITCH"] = (-555 + hot_robot_pitcher["ZIP"] + buff) * (1 - (1/((hot_robot_pitcher["IPfloat"] + 110) / 100)))
        else:
            hot_robot_pitchers.at[i, "ZPITCH"] = (hot_robot_pitcher["ZERA"] + hot_robot_pitcher["ZIP"] + buff) * (1 - (1/((hot_robot_pitcher["IPfloat"] + 100) / 100)))
    hot_robot_pitchers = hot_robot_pitchers.drop(columns=['index', 'IPfloat', 'ERfloat', "ERAfloat", "ZERA", "ZIP"])
    
    return hot_robot_pitchers

#workbook = gc.open_by_key(spreadsheet_key)
#sheet = workbook.worksheet("Single Season Pitcher Stats")
#values = sheet.get_all_values()
#pitchers = pd.DataFrame(values[1:], columns = values[0])
#hot_robot = ["BAD BRICKS", "THUNDERBALLS", "PAPER ROCKS", "UNEMPLOYED"]
#commando = ["LANDFOWL", "BEIGE GUYS", "GREAT HAMSTERS", "GRAVITY JUNKERS", "DONKEY PLASMA"]
#zpitcher("8", commando, pitchers)
```

```python
#calculate the Z-BAT stat
def zhitter(season, teams, hitter_stats):    
    hot_robot_hitters = hitter_stats[(hitter_stats['TEAM'].isin(teams))]
    hot_robot_hitters = hot_robot_hitters[hot_robot_hitters["SEASON"] == season].reset_index()
    hot_robot_hitters["ABfloat"] = 1
    for i, hot_robot_hitter in hot_robot_hitters.iterrows():
        hot_robot_hitters.at[i, "ABfloat"] = float(hot_robot_hitter["AB"]) * 100
        
    hot_robot_hitters["Hfloat"] = 1
    for i, hot_robot_hitter in hot_robot_hitters.iterrows():
        hot_robot_hitters.at[i, "Hfloat"] = float(hot_robot_hitter["H"]) * 100
    hot_robot_hitters["TBfloat"] = 1
    for i, hot_robot_hitter in hot_robot_hitters.iterrows():
        hot_robot_hitters.at[i, "TBfloat"] = float(hot_robot_hitter["TB"]) * 100
        
    hot_robot_hitters["APSfloat"] = 1
    for i, hot_robot_hitter in hot_robot_hitters.iterrows():
        if hot_robot_hitter["AVG"] == "nan" or hot_robot_hitter["SLG"] == "nan" or hot_robot_hitter["AVG"] == "N/A" or hot_robot_hitter["SLG"] == "N/A":
            hot_robot_hitters.at[i, "APSfloat"] = 42069
        else:
            hot_robot_hitters.at[i, "APSfloat"] = (float(hot_robot_hitter["AVG"]) + float(hot_robot_hitter["SLG"])) * 100
    hot_robot_mean_ab = hot_robot_hitters["ABfloat"].median()
    max_ab = max(hot_robot_hitters["ABfloat"])
    hot_robot_sum_ab = hot_robot_hitters["ABfloat"].sum()
    hot_robot_sum_h = hot_robot_hitters["Hfloat"].sum()
    hot_robot_sum_tb = hot_robot_hitters["TBfloat"].sum()
    hot_robot_mean_aps = ((hot_robot_sum_h + hot_robot_sum_tb) / hot_robot_sum_ab) * 100
    hot_robot_stdev_ab = hot_robot_hitters["ABfloat"].std()    
    hot_robot_stdev_aps = era_stdev(hot_robot_mean_aps, hot_robot_hitters["APSfloat"].tolist())

    hot_robot_hitters["ZAPS"] = 0
    for i, hot_robot_hitter in hot_robot_hitters.iterrows():
        hot_robot_hitters.at[i, "ZAPS"] = ((hot_robot_hitter["APSfloat"] - hot_robot_mean_aps) / hot_robot_stdev_aps) * 100
    hot_robot_hitters["ZAB"] = 0
    for i, hot_robot_hitter in hot_robot_hitters.iterrows():
        hot_robot_hitters.at[i, "ZAB"] = ((hot_robot_hitter["ABfloat"] - hot_robot_mean_ab) / hot_robot_stdev_ab) * 50
    hot_robot_hitters["ZBAT"] = 0
    for i, hot_robot_hitter in hot_robot_hitters.iterrows():
        if hot_robot_hitter["AVG"] == "nan" or hot_robot_hitter["AVG"] == "N/A":
            hot_robot_hitters.at[i, "ZBAT"] = 0
        else:
            hot_robot_hitters.at[i, "ZBAT"] = ((hot_robot_hitter["ZAPS"])) * (hot_robot_hitter["ABfloat"] / max_ab)
    
    
    
    #buff = ((1 - hot_robot_mean_ab) / hot_robot_stdev_ab) * -25
    #for i, hot_robot_hitter in hot_robot_hitters.iterrows():
    #    if hot_robot_hitter["AVG"] == "nan":
    #        hot_robot_hitters.at[i, "ZBAT"] = 0
    #    else:
    #        hot_robot_hitters.at[i, "ZBAT"] = (hot_robot_hitter["ZAPS"] + hot_robot_hitter["ZAB"] + buff) * (1 - (1/((hot_robot_hitter["ABfloat"] + 100) / 100)))
    hot_robot_hitters = hot_robot_hitters.drop(columns=['index', 'ABfloat', 'Hfloat', 'TBfloat', "APSfloat", "ZAPS", "ZAB"])
    
    return hot_robot_hitters

#workbook = gc.open_by_key(spreadsheet_key)
#sheet = workbook.worksheet("Single Season Batter Stats")
#values = sheet.get_all_values()
#hitters = pd.DataFrame(values[1:], columns = values[0])
#hot_robot = ["BAD BRICKS", "THUNDERBALLS", "PAPER ROCKS", "UNEMPLOYED"]
#commando = ["DONKEY PLASMA", "BEIGE GUYS", "GREAT HAMSTERS", "GRAVITY JUNKERS", "LANDFOWL"]
#zhitter("2", commando, hitters)
```

```python
def z_hitter_catch_up():
    workbook = gc.open_by_key(spreadsheet_key)
    sheet = workbook.worksheet("Single Season Batter Stats")
    values = sheet.get_all_values()
    hitters = pd.DataFrame(values[1:], columns = values[0])
    hitters["SEASON"] = pd.to_numeric(hitters["SEASON"])
    hitters["AB"] = pd.to_numeric(hitters["AB"])
    hitters["H"] = pd.to_numeric(hitters["H"])
    hitters["TB"] = pd.to_numeric(hitters["TB"])

    for i, hitter in hitters.iterrows():
        if hitter["AB"] == 0:
            hitters.at[i, "AVG"] = "N/A"
            hitters.at[i, "SLG"] = "N/A"
        else:
            hitters.at[i, "AVG"] = (hitter["H"] / hitter["AB"])
            hitters.at[i, "SLG"] = (hitter["TB"] / hitter["AB"])

                                    
    season = 1
    hot_robot = ["BAD BRICKS", "THUNDERBALLS", "PAPER ROCKS", "UNEMPLOYED"]
    commando = ["DONKEY PLASMA", "BEIGE GUYS", "GREAT HAMSTERS", "GRAVITY JUNKERS", "LANDFOWL"]
    
    out = pd.DataFrame(columns = ["PLAYER", 'TEAM', 'SEASON', 'AB', 'H', 'TB', 'R', 'HR', 'RBI', 'SB', "AVG",
                                  "SLG", "ROTY_ELIG", "ALL-STAR", "DS-MVP", "WS-MVP", "ROTY (H)", "MVP (H)",
                                  "CHAMPIONSHIPS", "ZBAT"])
    
    while season <= hitters["SEASON"].max():
        hot_robot_season = zhitter(season, hot_robot, hitters).reset_index()
        commando_season = zhitter(season, commando, hitters).reset_index()
        season += 1
        out = out.append(hot_robot_season).append(commando_season)
        out = out.reset_index().drop(columns = ["level_0", "index"])
        out = out[["PLAYER", 'TEAM', 'SEASON', 'AB', 'H', 'TB', 'R', 'HR', 'RBI', 'SB', "AVG", 
                  "SLG", "ZBAT", "ROTY_ELIG", "ALL-STAR", "DS-MVP", "WS-MVP", "ROTY (H)", "MVP (H)",
                  "CHAMPIONSHIPS"]]
    d2g.upload(out, spreadsheet_key, "Single Season Batter Stats", credentials=credentials, row_names=False)

#z_hitter_catch_up()
```
